### Original Code on diabetes dataset

```python
import keras
import pandas
from keras.models import Sequential
from keras.layers import Dense, Activation

# load dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dataset = pd.read_csv('diabetes.csv', header=None).values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=8, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

### After adding more layers


```python
dataset = pd.read_csv('diabetes.csv', header=None).values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
network2 = Sequential() # create model
network2.add(Dense(35, input_dim=8, activation='sigmoid')) # hidden layer
network2.add(Dense(30, input_dim=8, activation='sigmoid')) # hidden layer
network2.add(Dense(10, activation='relu')) # hidden layer
network2.add(Dense(5, activation='relu')) # hidden layer
network2.add(Dense(1, activation='sigmoid')) # output layer
network2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
network2_fitted = network2.fit(X_train, Y_train, epochs=150,
                                     initial_epoch=0)
print(network2.summary())
print(network2.evaluate(X_test, Y_test))
```

### Breast Cancer data set - pre normalization

```python
data = pd.read_csv('breastcancer.csv')

def get_target(x):
    return 1 if x == 'M' else 0

data['target'] = data['diagnosis'].apply(get_target)
X = data.iloc[:, 2:32]
y = data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y,
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
network1 = Sequential() # create model
network1.add(Dense(40, input_dim=30, activation='sigmoid')) # hidden layer
network1.add(Dense(20, activation='relu')) # hidden layer
network1.add(Dense(10, activation='relu')) # hidden layer
network1.add(Dense(15, activation='relu')) # hidden layer
network1.add(Dense(1, activation='sigmoid')) # output layer
network1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
network1_fitted = network1.fit(X_train, Y_train, epochs=150,
                                     initial_epoch=0)
print(network1.summary())

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

Y_pred_prob = network1.predict(X_test)
Y_pred = np.where(Y_pred_prob > 0.5, 1, 0)

cm = confusion_matrix(Y_test, Y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```
