```python
import pandas as pd
df = pd.read_parquet('games.parquet')
df
```

```python
df = df[['MatchID', 'TeamID', 'ChemDragDiff15','CloudDragDiff15', 'HexDragDiff15', 'InfDragDiff15', 'MountainDragDiff15','OceanDragDiff15',
    'GrubDiff','GoldDiff15','XPDiff15','WonGame', 'Rank', 'Region', 'Patch']]
df
```

```python
import re 
df['Patch'] = df['Patch'].str.replace(r'^([^\.]*\.[^\.]*)\..*$', r'\1', regex=True)
df
```

```python
def patch_check(s):
    parts =  s.split('.')
    if int(parts[1]) > 8:
        return True
    else:
        return False
```

```python
df['keep'] = df['Patch'].apply(patch_check)
df
```

```python
df = df[df['keep']]
df['Patch'].value_counts()
```

```python
df['total_drag_diff15'] = df['ChemDragDiff15'] + df['CloudDragDiff15'] + df['HexDragDiff15'] + df['InfDragDiff15'] + df['MountainDragDiff15'] + df['OceanDragDiff15']
df
```

### Grub Chi-Square

```python
from scipy.stats import chi2_contingency
import numpy as np

table = np.array([
    [673, 620],  # -6
    [14, 23],    # -5
    [321, 352],  # -4
    [120, 140],  # -3
    [336, 376],  # -2
    [78,68],     # -1
    [880,886],   # 0
    [85,80],     # 1
    [348,351],   # 2
    [122, 125],  # 3
    [337, 331],  # 4 
    [23, 22],    # 5
    [667, 622],  # 6
])

chi2, p_value, dof, expected = chi2_contingency(table)
print('x^2:', chi2)
print('p: ', p_value)

V = np.sqrt(chi2 / 7980)

print('V:', V)
```

### Dragon Chi-Square


```python
table = np.array([
    [335, 533],  # -2
    [1019,1336], # -1
    [791,731],    #  0
    [1327,1028],  #  1
    [532,348],    #  2
])

chi2, p_value, dof, expected = chi2_contingency(table)
print('x^2:', chi2)
print('p: ', p_value)

V = np.sqrt(chi2 / 7980)

print('V:', V)
```

No evidence that winning is dependent on grub difference

Evidence that winning is dependent on dragon difference

### Grub Z-Test

```python
from statsmodels.stats.proportion import proportions_ztest

wins = [673,14,321,120,336,78,880,85,348,122,337,23,667]
nobs = [1293, 37, 673, 260, 712, 146, 1746, 165, 699, 247, 668, 45, 1289]

for i in range(len(wins)):
    z, p = proportions_ztest(wins[i], nobs[i], value=.5)
    print(p)
```

### Dragon Z-Test

```python
wins = [335, 1019, 791, 1327, 532]
nobs = [868, 2355, 1522, 2355, 880]

for i in range(len(wins)):
    z, p = proportions_ztest(wins[i], nobs[i], value=.5)
    print(p)
```

No evidence that the groups with grub differences have a winrate that is different from .5

Evidence that the groups with dragon differences have a winrate that is different from .5

### SHAP

```python
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
import shap

X = df[['total_drag_diff15', 'GrubDiff', 'XPDiff15', 'GoldDiff15']]
y = df['WonGame']

model = CatBoostClassifier()
model.fit(X,y, verbose=False)
```

```python
data = Pool(X,y)
shapvalues = model.get_feature_importance(data,
                             type='ShapValues',
                             verbose=False)

mean_abs_shap_values = np.abs(shapvalues).mean(axis=0)

mean_abs_shap_values
```

```python
feature_names = X.columns.append(pd.Index(['intercept']))
shap_summary = pd.DataFrame({'Feature': feature_names, 'Mean Abs SHAP Value': mean_abs_shap_values})
shap_summary = shap_summary.sort_values(by='Mean Abs SHAP Value', ascending=False)

print(shap_summary)
```
